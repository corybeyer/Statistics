Suppose another Y observation is obtained at X = 45. Would E{Y} for this new observation still be 104? Would the Y value for this new case again be 108?



```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import math
```

```python
def equ(x):
    return 20*x+100
x = np.linspace(0,5,5)
```

```python
fig, ax = plt.subplots()
ax.plot(x,equ(x))
ax.plot(5, 200, 'o', color='black')
ax.text(.015, .85, r'$y = mx+b$', {'fontsize': 13},transform=ax.transAxes)
ax.text(.015, .77, r'$y = 20x+100$', {'fontsize': 13},transform=ax.transAxes)
```

```python
fig, ax = plt.subplots()
mu = 200
variance = 25
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.text(.015, .85, r'$Y = \beta_0+\beta_0X_i$', {'fontsize': 10},transform=ax.transAxes)
plt.text(.015, .77, r'$X_1 =5, Y_1 =  200$', {'fontsize': 10},transform=ax.transAxes)
plt.text(.015, .69, r'$\beta_0 = 100, \beta_1 =  20, \sigma = 5$', {'fontsize': 10},transform=ax.transAxes)
plt.text(.015, .61, r'$Y = 100 + 20(5)= 200$', {'fontsize': 10},transform=ax.transAxes)
plt.axvline(x=mu)

probs =stats.norm(loc=200, scale = 5).cdf(205)-stats.norm(loc=200, scale = 5).cdf(195)
plt.text(.75, .85, np.round(probs,2), {'fontsize': 10},transform=ax.transAxes)

section = np.arange(195, 206, 1)
plt.fill_between(section,stats.norm.pdf(section, mu, sigma))
plt.show()
```
